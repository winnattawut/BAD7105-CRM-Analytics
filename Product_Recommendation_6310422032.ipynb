```
import pandas as pd

df = pd.read_csv('Customer Preference Survey (Responses) - Form Responses 1.csv')
```

```
df.head(5)
```

```
missing = df.isnull().sum()
print(missing)
```

```
df = df.fillna("ไม่")
```

```
def encode_units(x):
    if "ไม่" in str(x) :
      return 0
    elif "ไม่" not in str(x) :
      return 1

hot_encoded_df = df.applymap(encode_units)
```

```
df.isnull().sum()
```

```
hot_encoded_df.head(5)
```

```
hot_encoded_df.drop('Timestamp', axis=1, inplace=True)
```

```
hot_encoded_df.head(5)
```

```
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

frequent_itemsets = apriori(hot_encoded_df, min_support=0.02, use_colnames=True, max_len=3)
```

```
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head(10)
```

```
rules.shape
```

```
rule =  rules[ (rules['lift'] >= 1) & (rules['confidence'] >= 0.5) &  (rules['support'] >= 0.2) ]
rule
```

```
market_basket = rule.sort_values(by=['lift', 'confidence', 'support', 'consequent support', 'antecedent support'], ascending=False)
market_basket.head(10)
```
